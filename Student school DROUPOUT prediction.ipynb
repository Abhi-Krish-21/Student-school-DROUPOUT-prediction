```python
#creating a proxy droupout risk indicator
df["low_grade"] = ((df["G1"]<8) | (df["G2"]<8) | (df["G3"]<8)).astype(int)
df["high_absence"] = (df["absences"] > 50).astype(int)
df["low_study_time"] = (df["studytime"]<=2).astype(int)
df["high_failure"] = (df["failures"] == 3).astype(int) 

#family issues
df['parents_apart'] = (df['Pstatus_T'] == 0).astype(int)
df['low_parental_education'] = ((df['Medu'] <= 2) | (df['Fedu'] <= 2)).astype(int)
df['low_family_support'] = (df['famsup_yes'] == 0).astype(int)
df['parent_unemployment'] = ((df['Mjob_home'] == 1) | (df['Fjob_home'] == 1)).astype(int)

#combining all family issues into a single feature
df['family_issues'] = df[['parents_apart', 'low_parental_education', 'low_family_support', 'parent_unemployment']].sum(axis=1)
```

```python
df["risk_score"] = (df[["low_grade", "high_absence", "low_study_time", "high_failure", 'family_issues']]).sum(axis=1)

def levels_of_risk(score):
    if score >= 4:
        return "high"
    else:
        return "low"
    
df['risk_level'] = df['risk_score'].apply(levels_of_risk)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.countplot(x='risk_level', data=df, palette='viridis')
plt.title('Distribution of Students by Risk Level')
plt.xlabel('Risk Level')
plt.ylabel('Number of Students')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='risk_level', y='G3', data=df, palette='viridis')
plt.title('Final Grades by Risk Level')
plt.xlabel('Risk Level')
plt.ylabel('Final Grade (G3)')
plt.show()
```

```python
#clustering students based on their education & behaviour
from sklearn.cluster import KMeans

features = ["G1", "G2", "G3", "absences", "studytime"]
X = df[features]
kmeans = KMeans(n_clusters = 3, random_state = 42)
clusters = kmeans.fit_predict(X)

df["Cluster"] = clusters

plt.figure(figsize = (10,10))
sns.scatterplot(x=df["G1"], y=df["G3"], hue = df["Cluster"], palette = "viridis")
plt.show()
```

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

target = "risk_level"
feature = df1.drop(columns = ["Cluster", target]).columns

x = df1[feature]
y = df1[target]
x1, x2, y1, y2 = tts(x, y, test_size = 0.3, random_state = 42)

model = RFC(random_state = 42)
model.fit(x1, y1)

y_pred = model.predict(x2)
print(confusion_matrix(y2, y_pred))
print(classification_report(y2, y_pred))
print('Accuracy:', accuracy_score(y2, y_pred))

predicted_risk_level = y_pred[0:119]
print(f"The predicted dropout risk is: {predicted_risk_level}")
```
